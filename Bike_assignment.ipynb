```python
bikedata['yr'] = bikedata['yr'].replace({0:'2018',1:'2019'})
bikedata.tail()
```

```python
bikedata['season'] = bikedata['season'].replace({1:'spring', 2:'summer', 3:'fall', 4:'winter'})
bikedata.head()
```

```python
bikedata['mnth'] = bikedata['mnth'].replace({1:'January', 2:'February', 3:'March', 4:'April', 5:'May', 6:'June', 7:'July', 8:'August', 9:'September', 10:'October', 11:'November', 12:'December'})
bikedata.head()
```

```python
bikedata['weathersit'] = bikedata['weathersit'].replace({1:'Clear, Few clouds, Partly cloudy, Partly cloudy', 2:'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist', 3:'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds', 4:'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'})
bikedata.head()
```

```python
bikedata['weekday'] = bikedata['weekday'].replace({0: 'Sunday', 1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday'})
bikedata.head()
```

```python
# percentage of missing values in each column
round(100*(bikedata.isnull().sum()/len(bikedata)), 2).sort_values(ascending=False)
```

```python
# percentage of missing values in each row
round((bikedata.isnull().sum(axis=1)/len(bikedata))*100,2).sort_values(ascending=True)
```

```python
#let's check for duplicates
bike_dup = bikedata.copy()

# Checking for duplicates and dropping the entire duplicate row if any
bike_dup.drop_duplicates(subset=None, inplace=True)
bike_dup.shape
```

```python
#instant and date are just sequences, and we are focusing on only one target variable 'cnt', so 'casual' and 'registered
#are redundant,so we can remove them
bike_testdata = bikedata.drop(['instant', 'dteday', 'casual','registered' ], axis=1)
bike_testdata.shape
```

```python
#season,yr,mnth,weekday,weathersit as we have changed their values, and they are actually categorical in nature
#so let's change them in categorical variables
#bike_testdata['yr']=bike_testdata['yr'].astype('category')
bike_testdata['mnth']=bike_testdata['mnth'].astype('category')
bike_testdata['season']=bike_testdata['season'].astype('category')
bike_testdata['weekday']=bike_testdata['weekday'].astype('category')
bike_testdata['weathersit']=bike_testdata['weathersit'].astype('category')

bike_testdata.info()

```

Building a Linear Model Now

```python
#dividing in x and y sets
y_train = df_train.pop('cnt')
x_train = df_train
```

```python
# Importing RFE(Recursive Feature Elimination) and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE 
lm = LinearRegression()
lm.fit(x_train, y_train)
 # running RFE
rfe = RFE(lm)           
rfe = rfe.fit(x_train, y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

```python
x_train.columns[~rfe.support_]
```

```python
# Creating X_train dataframe with RFE selected variables
X_train_rfe = x_train[col]
X_train_rfe.head()
```

```python
#Building Model using stats
#VIF
#Check for the VIF values of the feature variables.
#Model1 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = True)
vif
```

```python
import statsmodels.api as sm

# Add a constant
X_train_m1 = sm.add_constant(X_train_rfe)

# Create a first fitted model
model1 = sm.OLS(y_train, X_train_m1).fit()
# Check the parameters obtained
model1.params
```

```python
# Print a summary of the linear regression model obtained
print(model1.summary())
```
